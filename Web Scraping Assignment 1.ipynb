```python
import pandas as pd
def IMDB(website):
    webpage=requests.get(website)
    soup=BeautifulSoup(webpage.content)
        
    #Scrap the movie names
    movie_names=[]
    for i in soup.find_all('td',class_='titleColumn'):
        movie_names.append(i.text.split('\n')[2].strip(' '))
    movie_names
    
    #Scrap the release year
    year_release=[]
    for j in soup.find_all('span',class_='secondaryInfo'):
        year_release.append(j.text.strip('()'))
    year_release
    
    #Scrap the rating
    rating=[]
    for k in soup.find_all('td',class_='ratingColumn imdbRating'):
        rating.append(k.text.strip('\n'))
    rating
    
    #Store the Scrapped data in a Dataframe
    IMDB_Top_Movies=pd.DataFrame()
    IMDB_Top_Movies['Movie_Name']=movie_names
    IMDB_Top_Movies['Release_Year']=year_release
    IMDB_Top_Movies['IMDB_Rating']=rating
    IMDB_Top_Movies.index=IMDB_Top_Movies.index+1
    IMDB_Top_Movies.index.name='Ranking'
    
    #Creating Dataframe with top 100 movies
    IMDB_Top100Movies=IMDB_Top_Movies.head(100)
    return(IMDB_Top100Movies)

    
```

```python
IMDB('https://www.imdb.com/chart/top/')
```

```python
def ICC_Men(website):
    webpage=requests.get(website)
    soup=BeautifulSoup(webpage.content,'lxml')
    
    #Scrap ODI Team Rankings
    #Scrap table rows to know the row tags
    trows=soup.find_all('tr')[1:]
    trows
    
    #Scrap the row content tags to select the desired data
    trd=[]
    for i in soup.find_all('tr')[1:]:
        trd.append(i.find_all('td'))
    trd
    
    #Scrap top 10 team names
    team=[trd[i][1].text.split('\n')[2] for i in range(10)]
    
    #Scrap no of matches data
    matches=[trd[i][2].text for i in range(10)]
    
    #Scrap Points data
    points=[trd[i][3].text for i in range(10)]
     
    #Scrap Ratings data
    rating=[int(trd[i][4].text) for i in range(10)]
    
    #Store the Scrapped data in a DataFrame
    ICC_Men_ODI_Team=pd.DataFrame()
    ICC_Men_ODI_Team['Team']=team
    ICC_Men_ODI_Team['Matches']=matches
    ICC_Men_ODI_Team['Points']=points
    ICC_Men_ODI_Team['Rating']=rating
    ICC_Men_ODI_Team.index=ICC_Men_ODI_Team.index+1
    ICC_Men_ODI_Team.index.name='Position'
    print(ICC_Men_ODI_Team)
```

```python
ICC_Men('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
```

```python
def Nobroker(website):
    webpage=requests.get(website)
    soup=BeautifulSoup(webpage.content,'lxml')
    
    #Scrap House details
    #Scrap House Title
    title=[t.text for t in soup.find_all('h2', class_='heading-6 font-semi-bold nb__1AShY')]
    
    #Scrap Location
    location=[l.text.split(',')[1:] for l in soup.find_all('div',class_='nb__2CMjv')]
    
    #Scrap Area details
    area=[a.text.strip('sqft') for a in soup.find_all('div',class_='nb__3oNyC')]
    
    #Scrap EMI
    emi=[e.text for e in soup.find_all('div',id='roomType')]
    
    #Scrap Price
    price=[p1.text.split('₹')[1] for p1 in soup.find_all('div',id='minDeposit')]
    
    #Store House details in a DataFrame
    House_details=pd.DataFrame()
    House_details['Title']=title
    House_details['Location']=location
    House_details['Area in sqft']=area
    House_details['EMI']=emi
    House_details['Price in Rupees']=price
    print(House_details)
```

```python
def Dineout(website):
    webpage=requests.get(website)
    soup=BeautifulSoup(webpage.content,'lxml')
    
    #Scrap Restaurant names
    names=[n.text for n in soup.find_all('a',class_='restnt-name ellipsis')]
    
    #Scrap Cuisine details
    cuisine=[c.text.split('|')[1] for c in soup.find_all('div',class_='detail-info')]
    
    #Scrap Location details
    location=[l.text for l in soup.find_all('div',class_='restnt-loc ellipsis')]
    
    #Scrap Restaurant rating
    r1=soup.find('div',class_='restnt-rating rating-3')
    r1=r1.text
    
    rating=[r1]
    for r in soup.find_all('div',class_='restnt-rating rating-4') :
        rating.append(r.text)

    #Scrap Image URL
    url=[i['data-src'] for i in soup.find_all('img',class_='no-img')]
    
    #Store restaurant details in a DataFrame
    Dineout_Delhi=pd.DataFrame()
    Dineout_Delhi['Restaurant Name']=names
    Dineout_Delhi['Cuisine']=cuisine
    Dineout_Delhi['Location']=location
    Dineout_Delhi['Rating']=rating
    Dineout_Delhi['Image URL']=url
    print(Dineout_Delhi)
    
```

```python
Dineout('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```
