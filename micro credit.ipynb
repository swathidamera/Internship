# Data Cleaning

There are some customers with no loan history which can be known from 'cnt_loans90' column. They are not useful for the analysis. Remove them

```python
len(df1[df1['cnt_loans90']==0])
```

```python
for i,c in enumerate(df1['cnt_loans90']):
    if c==0.0:
        df1.drop(i,inplace=True)
```

```python
df1.reset_index(drop=True,inplace=True)
```

It can be seen that there are decimal values and maximum value around 5000 which is not possible. 

```python

for i,c in enumerate(df1['cnt_loans90']):
    if str(c).split('.')[1] != '0':       #identifies the values with decimals
        
        #if maximum loan in the last 90days is 6,all the loans will be of 6.So no of loans=total amount of loans /6
        if (df1['maxamnt_loans90'][i]==6):
            df1['cnt_loans90'][i]=df1['amnt_loans90'][i]/6   
        
        elif (df1['maxamnt_loans90'][i]== df1['maxamnt_loans30'][i] and df1['amnt_loans90'][i]== df1['amnt_loans30'][i]):
            df1['cnt_loans90'][i]=df1['cnt_loans30'][i]
            
        else:
            df1.drop(i,inplace=True)
            
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
df1.describe()
```

'maxamnt_loans30' can have 6 or 12 and no other values. Other values in this column have to be treated.
It is observed that other values have decimals. Using this fact identify the values other than 6 or 12

```python
for i,c in enumerate(df1['maxamnt_loans30']):
    if str(c).split('.')[1] != '0': #identifies the values with decimals
        if (df1['cnt_loans90'][i]== df1['cnt_loans30'][i] and df1['amnt_loans90'][i]== df1['amnt_loans30'][i]):
            df1['maxamnt_loans30'][i]=df1['maxamnt_loans90'][i]
                    
        elif (df1['amnt_loans30'][i] % df1['cnt_loans30'][i])==0 :
            df1['maxamnt_loans30'][i]= df1['amnt_loans30'][i]/df1['cnt_loans30'][i]
            
        else:
            df1.drop(i,inplace=True)
            
```

# Fitting Models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_curve,roc_auc_score
from sklearn.model_selection import cross_val_score
```

```python
y.value_counts(normalize=True)
```

There is imbalance in the target variable. So split the data in a stratified way. Use f1Score and ROC-AUC curve to choose the best fit model.

# Logistic Regression

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0,stratify=y)
lgr=LogisticRegression()
lgr.fit(x_train,y_train)
pred=lgr.predict(x_test)
print('accuracy score',accuracy_score(y_test,pred))
print('confusion matrix   ',confusion_matrix(y_test,pred))
print('classification report')
print(classification_report(y_test,pred))
```

# AUC- ROC Curve

```python
y_pred_prob=lgr.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='lgr')
plt.xlabel('False Positive Rate')
plt.ylabel('True positive rate')
plt.title('LGR')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,lgr.predict(x_test))
auc_score
```

```python
score=cross_val_score(lgr,x,y,cv=8)
print(score.mean())
```

# Decision Tree Classifier

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print('accuracy score',accuracy_score(y_test,pred))
print('confusion matrix',confusion_matrix(y_test,pred))
print('classification report',classification_report(y_test,pred))
```

# AUC-ROC Curve

```python
y_pred_prob=dtc.predict_proba(x_test)[:,1]
fpr,tpr,thresholds=roc_curve(y_test,y_pred_prob)
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='dtc')
plt.xlabel('False Positive Rate')
plt.ylabel('True positive rate')
plt.title('dtc')
plt.show()
```
