Question 8

```python
#Get the URL
driver.get('https://www.forbes.com/billionaires/')
```

```python
#Maximise the window
driver.maximize_window()
```

```python
Rank=[]
Name=[]
Networth=[]
Age=[]
Country=[]
Source=[]
Category=[]
#Extract the data of each column from the page

for x in range(20):
    Rank.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='rank']")])
    Name.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='personName']")])
    Networth.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='netWorth']")])
    Age.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='age']")])
    Country.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='countryOfCitizenship']")])
    Source.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='source']")])
    Category.extend([i.text for i in driver.find_elements_by_xpath("//div[@class='category']")])

    try:
        driver.find_element_by_xpath("//div[@class='next-page']").click()
        time.sleep(0.5)
    except ElementClickInterceptedException:
        print('Done')
        break
```

```python
#Create a Dataframe 'Billionaires' to store the details
Billionaires=pd.DataFrame()
Billionaires['Rank']=Rank
Billionaires['Name']=Name
Billionaires['Networth']=Networth
Billionaires['Age']=Age
Billionaires['Country']=Country
Billionaires['Source']=Source
Billionaires['Category']=Category
```

```python
pd.set_option('display.max_rows', None)
Billionaires
```
