```python
containers=driver.find_elements_by_xpath("//div[@class='lister-item-content']")
```

```python
name=[]
year=[]
genre=[]
runtime=[]
ratings=[]
votings=[]
for i in containers:
    
    #Name
    try:
        name.append(i.find_element_by_xpath(".//h3/a").text)
    except NoSuchElementException:
        name.append('--')
        
    #Year span
    try:
        year.append(i.find_element_by_xpath(".//span[@class='lister-item-year text-muted unbold']").text)
    except NoSuchElementException:
        year.append('--')
        
    #Genre
    try:
        genre.append(i.find_element_by_xpath(".//p/span[@class='genre']").text)
    except NoSuchElementException:
        genre.append('--')
    
    #Run time
    try:
        runtime.append(i.find_element_by_xpath(".//p/span[@class='runtime']").text)
    except NoSuchElementException:
        runtime.append('--')
    
    #Ratings
    try:
        ratings.append(i.find_element_by_xpath(".//div/div/span[2]").text)
    except NoSuchElementException:
        ratings.append('--')
        
    #Votings
    try:
        votings.append(i.find_element_by_xpath(".//span[@name='nv']").text)
    except NoSuchElementException:
        votings.append('--')
        
```

```python
Tv_Series=pd.DataFrame()
Tv_Series['Name']=name
Tv_Series['Year Span']=year
Tv_Series['Genre']=genre
Tv_Series['Runtime']=runtime
Tv_Series['Rating']=ratings
Tv_Series['Votes']=votings
Tv_Series.index=Tv_Series.index+1
Tv_Series
```

```python
Tv_Series.to_csv('Tv_Series')
```
