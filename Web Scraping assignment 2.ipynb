```python
#switch to tab browser
p= driver.window_handles[0]
c = driver.window_handles[1]
driver.switch_to.window(c)
```

```python
#Scrap Job info 
info=driver.find_elements_by_xpath("//div[@class='company-info']")
name=[i.text.strip('exp').split('\n')[0] for i in info]
salrecord=[i.text.strip('exp').split('\n')[1] for i in info]
exp=[i.text.strip('exp').split('\n')[-1] for i in info]
```

```python
#Scrap Salary info
salary=driver.find_elements_by_xpath("//div[@class='result-col salary-range']")
avgsal=[i.text.split('\n')[0] for i in salary ]
minsal=[i.text.split('\n')[1] for i in salary ]
maxsal=[i.text.split('\n')[2] for i in salary ]
```

```python
#Dataframe
Salarydata=pd.DataFrame()
Salarydata['Company Name']=name
Salarydata['Total Salary Record']=salrecord
Salarydata['Experience Required']=exp
Salarydata['Average Salary']=avgsal
Salarydata['Minimum Salary']=minsal
Salarydata['Maximum Salary']=maxsal
Salarydata.index=Salarydata.index+1
Salarydata
```
