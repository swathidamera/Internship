```python
#Get the URL
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Maximise the window
driver.maximize_window()
```

```python
#Locate the table with list of most viewed Youtube videos and find the rows containing the contents
rank=[i.text for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")]
name=[i.text.split('[')[0].strip('"') for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")]
uploader=[i.text for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")]
views=[i.text for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")]
date=[i.text for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]")]
```

```python
#Store in a dataframe the scraped contents
Youtube_Mostviewed=pd.DataFrame()
Youtube_Mostviewed['Rank']=rank
Youtube_Mostviewed['Video Name']=name
Youtube_Mostviewed['Artist']=uploader
Youtube_Mostviewed['Views(billions)']=views
Youtube_Mostviewed['Upload Date']=date
```

```python
Youtube_Mostviewed.to_csv('Youtube_Mostviewed',index=False)
```

```python
from selenium.webdriver.support.ui import Select
```

```python
#Get the URL
driver.get('https://www.bcci.tv/')
```

```python
#Maximise the window
driver.maximize_window()
```

```python
#Click on the Menu icon
driver.find_element_by_xpath("//span[@class='menu-icon__line menu-icon__line-right']").click()
```

```python
#Find the list of items in Menu
menu_items=driver.find_elements_by_xpath("//li[contains(@class,'nav-item')]")
```

```python
#Iterate over the menu items and click on International
for item in menu_items:
    if item.text=='INTERNATIONAL':
        item.click()
        break
```

```python
#Scrap the details of each International fixture 
series=[i.text for i in driver.find_elements_by_xpath("//span[@class='ng-binding']")]
match_title=[i.text.split('-')[0]  for i in driver.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")]
date=[i.text for i in driver.find_elements_by_xpath("//h5[@class='ng-binding']")]
time=[i.text for i in driver.find_elements_by_xpath("//h5[@class='text-right ng-binding']")]
place=[i.text.split('-')[1:]  for i in driver.find_elements_by_xpath("//div[@class='fix-place ng-binding ng-scope']")]
```

```python
#Create a DataFrame and store the details in the DataFrame
International_Fixtures=pd.DataFrame()
International_Fixtures['Series']=series
International_Fixtures['Match_Title']=match_title
International_Fixtures['Date']=date
International_Fixtures['Time']=time
International_Fixtures['Place']=place
International_Fixtures.index=International_Fixtures.index+1
```

```python
International_Fixtures
```

```python
International_Fixtures.to_csv('International_Fixtures',index=False)
```

```python
#Get the webpage
driver.get('https://www.guru99.com/')
```

```python
#Maximise the window
driver.maximize_window()
```

```python
#Find the element for 'Selenium' under Testing and click
try:
    driver.find_element_by_link_text('Selenium').click()
    time.sleep(1)
except NoSuchElementException:
    pass
```

```python
#Find the element for 'Selenium Exception Handling' and click
try:
    driver.find_element_by_xpath("//a[starts-with(@title,'Selenium Exception')]").click()
    time.sleep(1)
except NoSuchElementException:
    pass
    
```

```python
#Collect the elements of rows content of the table having 'Selenium Exception Handling' details
trows=driver.find_elements_by_xpath("//table/tbody/tr/td")
```

```python
#Iterate over the trows and store the name and description of the exceptions in the lists
name=[]
description=[]
for i,ele in enumerate(trows,1):
    if i%2==0:
        try:
            description.append(ele.text)
        except:
            print('--')
    else:
        try:
            name.append(ele.text)
        except:
            print('--')
     
        
        
```

```python
#Store the data in a Dataframe Selenium_Exceptions
Selenium_Exceptions=pd.DataFrame()
Selenium_Exceptions['Exception Name']=name
Selenium_Exceptions['Description']=description
```

```python
pd.options.display.max_colwidth=500
Selenium_Exceptions.index=Selenium_Exceptions.index+1
Selenium_Exceptions
```

```python
Selenium_Exceptions.to_csv('Selenium_Exceptions')
```

```python
from selenium.webdriver.common.action_chains import ActionChains
action=ActionChains(driver)
```

```python
#Collect the menu elements where 'Economy' is present
try:
    options=driver.find_elements_by_class_name('dropbtn')
except NoSuchElementException:
    pass

#Iterate over these elements and find the element of 'Economy'
for i in options:
    if i.text=='Economy':
        economy_ele=i
```

```python
#Move over to 'Economy' ,find element for "India" and click
try:
    action.move_to_element(economy_ele).perform()
except NoSuchElementException:
    pass

try:
    driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/div/a[3]').click()
    time.sleep(1)
except NoSuchElementException:
    pass
```

```python
#Click on 'GDP of Indian states'
try:
    driver.find_element_by_link_text('» GDP of Indian states').click()
    time.sleep(1)
except  NoSuchElementException:
    pass
```

```python
#Collect the elements of the contents of the table into trows
try:
    trows=driver.find_elements_by_xpath("//table[@id='table_id']/tbody/tr/td")
except NoSuchElementException:
    pass
```

```python
#Extract the rows text into respective columns lists 
Rank=[i.text for i in trows[0:len(trows)+1:8]]
State=[i.text for i in trows[1:len(trows)+1:8]]
GSDP_CP_19_20=[i.text for i in trows[2:len(trows)+1:8]]
GSDP_CP_18_19=[i.text for i in trows[3:len(trows)+1:8]]
Share=[i.text for i in trows[4:len(trows)+1:8]]
GDP_billiondollars=[i.text for i in trows[5:len(trows)+1:8]]
GSDP_11_12P_19_20=[i.text for i in trows[6:len(trows)+1:8]]
GSDP_11_12P_18_19=[i.text for i in trows[7:len(trows)+1:8]]  
```

```python
#Make a DataFrame and store the details in the dataframe
GDP_India_States=pd.DataFrame()
GDP_India_States['Rank']=Rank
GDP_India_States['State']=State
GDP_India_States['GSDP_CP_19_20']=GSDP_CP_19_20
GDP_India_States['GSDP_CP_18_19']=GSDP_CP_18_19
GDP_India_States['Share']=Share
GDP_India_States['GDP_billiondollars']=GDP_billiondollars
GDP_India_States['GSDP_11_12P_19_20']=GSDP_11_12P_19_20
GDP_India_States['GSDP_11_12P_18_19']=GSDP_11_12P_18_19
```

```python
#Get the webpage
driver.get('https://www.github.com/')
```

```python
#Maximise the window
driver.maximize_window()
```

```python
#Find and click on Explore
driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()
```

```python
#Find and click on Trending
driver.find_element_by_link_text("Trending").click()
```

```python
#Find the elements of Title and extract the text
title=[i.text for i in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a")]
```

```python
#Get the urls
urls=[i.get_attribute('href') for i in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")]
```

```python
description=[]
count=[]
language=[]
for url in urls:
    driver.get(url)
    time.sleep(1)
    
    #Description
    try:
        description.append(driver.find_element_by_xpath("//p[@class='f4 my-3']").text)
    except NoSuchElementException:
        description.append('None')
        
    #Contributors Count
    try:
        count.append(driver.find_element_by_xpath("//a[contains(@href,'contributors')]/span").text)
    except NoSuchElementException:
        count.append('1')
        
   
```

```python
pd.options.display.max_colwidth=500
Trending_Repositaries=pd.DataFrame()
Trending_Repositaries['Title']=title
Trending_Repositaries['Description']=description
Trending_Repositaries['Contributors_count']=count
Trending_Repositaries
```
